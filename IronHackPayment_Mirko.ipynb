```python
date_col_request = ["created_at", "updated_at", "moderated_at", "reimbursement_date", "cash_request_received_date", "money_back_date", "send_at", "reco_creation", "reco_last_update"]
date_col_fees = [ "created_at", "updated_at", "paid_at", "from_date", "to_date"]

#converting the date columns to datetime and keeping the date
for col in date_col_request:
    df_request[col] = pd.to_datetime(df_request[col], errors='coerce').dt.date

for col in date_col_fees:
    df_fees[col] = pd.to_datetime(df_fees[col], errors='coerce').dt.date
```

```python
user_df = df_request.copy()
```

```python
df_request.columns
```

```python
user_df = user_df.drop(columns=['id', 'status', 'updated_at',
       'moderated_at', 'deleted_account_id', 'reimbursement_date',
       'cash_request_received_date', 'money_back_date', 'transfer_type',
       'send_at', 'recovery_status', 'reco_creation', 'reco_last_update'])
```

```python
user_df.sample(5)
```

```python
user_df['created_at'] = pd.to_datetime(user_df['created_at'])
```

```python
#aggregating to define the cohorts

first_requests = user_df.sort_values(by='created_at').drop_duplicates(subset='user_id', keep= 'first')
monthly_counts = first_requests['created_at'].dt.to_period('M').value_counts().sort_index()
print(monthly_counts)
```

```python
import matplotlib.pyplot as plt

monthly_counts.plot(kind='bar', figsize=(12, 6), title="First Cash Requests Per Cohort Month")
plt.xlabel("Cohort Month")
plt.ylabel("Number of First-Time Users")
plt.xticks(rotation=45)
plt.show()
```
